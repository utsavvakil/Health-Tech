```python
import csv
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
bart = pd.read_csv('bart.csv').as_matrix()
row = bart.shape[0]
mean_after_explosion = []
mean_after_no_explosion = []
num_explosions = []
range_pumps = []
std_deviation = []
for i in range(row):
    mean_after_explosion.append(bart[i][9])
    mean_after_no_explosion.append(bart[i][10])
    num_explosions.append(bart[i][5])
    range_pumps.append(bart[i][7])
    std_deviation.append(bart[i][8])
```

```python
plt.plot(mean_after_explosion,'g',label='Mean after Explosion')
plt.plot(mean_after_no_explosion,'coral',label='Mean after no Explosion')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Trial number')
plt.ylabel('Mean pumps')
plt.show()
```

```python
plt.plot(num_explosions,'g',label='Number of Explosions')
plt.plot(range_pumps,'coral',label='Range of Pumps')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Trial number')
plt.show()
```

```python
plt.plot(num_explosions,'g',label='Number of Explosions')
plt.plot(std_deviation,'coral',label='Standard deviation Pumps')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Trial number')
plt.show()
```

```python
gonogo = pd.read_csv('gonogo.csv').as_matrix()
```

```python
row = gonogo.shape[0]
correct_responses = []
incorrect_responses = []
mean_resp_time = []
mean_accuracy = []
date_time = []
for i in range(row):
    correct_responses.append(gonogo[i][3])
    incorrect_responses.append(gonogo[i][5])
    mean_resp_time.append(gonogo[i][8]/10)
    mean_accuracy.append(gonogo[i][7]*100)
    date_time.append(gonogo[i][1])
```

```python
plt.plot(mean_accuracy,'red',label='Mean Accuracy')
plt.plot(mean_resp_time,'blue',label='Mean response time')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel('Trial number')
plt.show()
```
